```python
import random
import math
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = 2.0 # รัศมี
N = 5001 # ทําการสุ่มค่าทั้งหมด 5,000 ครั้ง
d = {"Trials":[],"Pi":[]}
```

```python
for T in range(1,N):
  circle_p=0
  for i in range(T):
    x = random.uniform(-2.0, 2.0) #-2 ถึง รัศมี
    y = random.uniform(-2.0, 2.0) #-2 ถึง รัศมี
    
    x2 = x ** 2
    y2 = y ** 2

    if math.sqrt(x2 + y2) <= r:
      circle_p+=1

  d["Trials"].append(T)
  d["Pi"].append((circle_p/T)*4)
```

```python
# กราฟ scatter plot
df = pd.DataFrame(data=d)
plt.figure(figsize = (10,7))
plot = sns.scatterplot(x="Trials", y="Pi", s=30, marker="o", data=df)
plot.set(title='Monte Carlo Simulation to Estimate Value of Pi', xlabel="Number of Trials", ylabel="Value of Pi")
plt.axhline(y=3.14, color='r', linestyle='-')
plt.show()
```
