```python
import numpy as np
import pandas as pd
import scipy.stats as stats
df = pd.read_csv('BloodPressure.csv')
```

```python
df
```

```python
male = df.loc[df["sex"] =="Male"].describe()
male
```

```python
female = df.loc[df["sex"] =="Female"].describe()
female
```

```python
significance = 0.05
statM,p_valueM = stats.ttest_rel(male['bp_before'], male['bp_after'])
statF,p_valueF = stats.ttest_rel(female['bp_before'], female['bp_after'])
```

```python
print('t_stat: %0.4f, p_value: %0.4f' % (statM, p_valueM))
if p_valueM < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accpet the null hypothesis")
```

```python
print('t_stat: %0.4f, p_value: %0.4f' % (statF, p_valueF))
if p_valueF < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accpet the null hypothesis")
```
