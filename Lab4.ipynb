# Z-test
* Library คือ statsmodels `from statsmodels.stats.proportion import proportions_ztest`

image.png

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np
```

```python
# alpha value
significance = 0.05 
# ค่าจากตาราง
successes = np.array([320, 300])
samples = np.array([400, 425])
```

```python
# คำนวณ Z-test เเละP-value
stat,p_value = proportions_ztest(count=successes,nobs=samples,alternative='two-sided')
```

```python
print('z_stat: %0.5f, p_value: %0.5f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accept the null hypothesis")
```

# T-test
* Library คือ scipy `import scipy.stats as stats`

image.png

```python
import numpy as np
import scipy.stats as stats
```

```python
# alpha value
significance = 0.05
# ค่าจากตาราง
A = np.array([43, 53, 65, 49, 55, 60, 47, 50, 60, 55])
B = np.array([62, 43, 54, 67, 59, 45, 46, 63, 65, 45])
```

```python
# คำนวณ T-test เเละP-value
stat, p_value = stats.ttest_ind(a = A, b = B, equal_var = True)
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accpet the null hypothesis")
```

# Paired T-test
* Library คือ scipy `import scipy.stats as stats`

image.png

```python
import numpy as np
import scipy.stats as stats
```

```python
# alpha value
significance = 0.01
# ค่าจากตาราง
group1 = np.array([60, 45, 80, 87, 79, 75, 60, 30, 45])
group2 = np.array([75, 65, 90, 80, 89, 95, 85, 69, 40])
```

```python
# คำนวณ Paired T-test เเละP-value
stat, p_value = stats.ttest_rel(group1,group2)
```

```python
print('t_stat: %0.5f, p_value: %0.4f' % (stat, p_value))
if p_value < significance:
  print ("Reject the null hypothesis")
else:
  print ("Accpet the null hypothesis")
```

# Chi-Square Test
* Library คือ scipy `from scipy.stats import chi2_contingency`

image.png

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# ค่าจากตาราง
df = pd.DataFrame(index=["Married","Single"], data={'Male': [25, 35],'Female': [15,25]})
```

```python
# คำนวณ Chi-Square Test เเละP-value
chi2, p, dof, expected = chi2_contingency(df,correction=False)
```

```python
print(f"chi2 statistic: {chi2:.5g}")
print(f"p-value: {p:.5g}")
print(f"degrees of freedom: {dof}")
print("expected frequencies:")
print(expected)
```
